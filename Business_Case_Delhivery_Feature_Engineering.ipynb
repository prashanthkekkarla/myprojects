###Destination Name: Split and extract features out of destination. City-place-code (State)

```python
def location_name_to_state(x):
    l = x.split('(')
    if len(l) == 1:
        return l[0]
    else:
        return l[1].replace(')', "")
```

```python
def location_name_to_city(x):
    if 'location' in x:
        return 'unknown_city'
    else:
        l = x.split()[0].split('_')
        if 'CCU' in x:
            return 'Kolkata'
        elif 'MAA' in x.upper():
            return 'Chennai'
        elif ('HBR' in x.upper()) or ('BLR' in x.upper()):
            return 'Bengaluru'
        elif 'FBD' in x.upper():
            return 'Faridabad'
        elif 'BOM' in x.upper():
            return 'Mumbai'
        elif 'DEL' in x.upper():
            return 'Delhi'
        elif 'OK' in x.upper():
            return 'Delhi'
        elif 'GZB' in x.upper():
            return 'Ghaziabad'
        elif 'GGN' in x.upper():
            return 'Gurgaon'
        elif 'AMD' in x.upper():
            return 'Ahmedabad'
        elif 'CJB' in x.upper():
            return 'Coimbatore'
        elif 'HYD' in x.upper():
            return 'Hyderabad'
        return l[0]
```

```python
def location_name_to_place(x):
    if 'location' in x:
        return x
    elif 'HBR' in x:
        return 'HBR Layout PC'
    else:
        l = x.split()[0].split('_', 1)
        if len(l) == 1:
            return 'unknown_place'
        else:
            return l[1]
```

##3.In-depth analysis and feature engineering:

##To compare the difference between od_total_time and start_scan_to_end_scan, we can conduct a hypothesis test or perform visual analysis

**STEP-1: Set up Null Hypothesis**
* **Null Hypothesis ( H0 ):** The mean difference between od_total_time (Total Trip Time) and start_scan_to_end_scan (Expected total trip time) is equal to zero.
* **Alternate Hypothesis ( HA ):** The mean difference between od_total_time and start_scan_to_end_scan is not equal to zero.

**STEP-2: Checking for Basic Assumptions for the Hypothesis**
* **Distribution Check using QQ Plot:** Examine the QQ plot to assess whether the differences between the two variables follow a normal distribution.
* **Homogeneity of Variances using Levene's Test:** Perform Levene's test to check if the variances of od_total_time and start_scan_to_end_scan are equal.

**STEP-3: Define Test Statistics; Distribution of T under H0**
* If the assumptions of the t-test are met, proceed with the paired t-test.
* If assumptions are violated, consider non-parametric alternatives such as the Wilcoxon signed-rank test.

**STEP-4: Compute the P-value and Fix the Value of Alpha**
* Set the significance level (alpha) to 0.05.

**STEP-5: Compare P-value and Alpha**
* **Decision Rule:**
  * If p value < alpha : reject H0
  * If p-value > alpha : fail to reject H0

```python
delhivery2[['od_total_time', 'start_scan_to_end_scan']].describe()
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(delhivery2['od_total_time'], element = 'step', color = 'green')
sns.histplot(delhivery2['start_scan_to_end_scan'], element = 'step', color = 'pink')
plt.legend(['od_total_time', 'start_scan_to_end_scan'])
plt.plot()
```

```python
plt.figure(figsize = (15, 6))
plt.subplot(1, 2, 1)
plt.suptitle('QQ plots for od_total_time and start_scan_to_end_scan')
spy.probplot(delhivery2['od_total_time'], plot = plt, dist = 'norm')
plt.title('QQ plot for od_total_time')
plt.subplot(1, 2, 2)
spy.probplot(delhivery2['start_scan_to_end_scan'], plot = plt, dist = 'norm')
plt.title('QQ plot for start_scan_to_end_scan')
plt.plot()
```

####The visual inspection of the QQ plots suggests that the samples do not exhibit a normal distribution.

####Conducting the Shapiro-Wilk test for normality

* **Shapiro-Wilk Test for Normality:**

**Hypotheses:**
  * **Null Hypothesis (H0):** The sample follows a normal distribution.
  * **Alternative Hypothesis (HA):** The sample does not follow a normal distribution.

**Significance Level:**
  * alpha = 0.05

**Test Statistic:**
  * Shapiro-Wilk test for normality

```python
test_stat, p_value = spy.shapiro(delhivery2['od_total_time'].sample(5000))
print('p-value', p_value)
if p_value < 0.05:
    print('The sample does not follow normal distribution')
else:
    print('The sample follows normal distribution')
```

```python
test_stat, p_value = spy.shapiro(delhivery2['start_scan_to_end_scan'].sample(5000))
print('p-value', p_value)
if p_value < 0.05:
    print('The sample does not follow normal distribution')
else:
    print('The sample follows normal distribution')
```

```python
transformed_od_total_time = spy.boxcox(delhivery2['od_total_time'])[0]
test_stat, p_value = spy.shapiro(transformed_od_total_time)
print('p-value', p_value)
if p_value < 0.05:
    print('The sample does not follow normal distribution')
else:
    print('The sample follows normal distribution')
```

```python
transformed_start_scan_to_end_scan = spy.boxcox(delhivery2['start_scan_to_end_scan'])[0]
test_stat, p_value = spy.shapiro(transformed_start_scan_to_end_scan)
print('p-value', p_value)
if p_value < 0.05:
    print('The sample does not follow normal distribution')
else:
    print('The sample follows normal distribution')
```

####Levene's Test for Homogeneity of Variances

```python
# Null Hypothesis(H0) - Homogenous Variance

# Alternate Hypothesis(HA) - Non Homogenous Variance

test_stat, p_value = spy.levene(delhivery2['od_total_time'], delhivery2['start_scan_to_end_scan'])
print('p-value', p_value)
if p_value < 0.05:
    print('The samples do not have  Homogenous Variance')
else:
    print('The samples have Homogenous Variance ')
```

####Due to non-normal distribution, we'll use the Mann-Whitney U rank test, a non-parametric alternative to the T-Test for independent samples.

```python
test_stat, p_value = spy.mannwhitneyu(delhivery2['od_total_time'], delhivery2['start_scan_to_end_scan'])
print('P-value :',p_value)
```

* Since the p-value (> alpha), we fail to reject the null hypothesis. Therefore, there is insufficient evidence to conclude a significant difference between od_total_time and start_scan_to_end_scan.
